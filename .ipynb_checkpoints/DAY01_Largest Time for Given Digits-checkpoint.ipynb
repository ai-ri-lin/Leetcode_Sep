```python
'''
Given an array of 4 digits, return the largest 24 hour time that can be made.

The smallest 24 hour time is 00:00, and the largest is 23:59.  

Starting from 00:00, a time is larger if more time has elapsed since midnight.

Return the answer as a string of length 5.  If no valid time can be made, return an empty string.

Example 1:
Input: [1,2,3,4]
Output: "23:41"

Example 2:
Input: [5,5,5,5]
Output: ""

'''
class Solution(object):
    def time_twenty(self):
        self.first = 2
        self.A.remove(2)
        A = self.A
        second = -1
        for ind in range(len(A)-1,-1,-1):
            if (A[ind] < 4):
                second = A[ind]
            else:
                break   
                
        if second == -1:
            pass
        else:
            A.remove(second)
            self.second = second
            self.time_minutes()
    
    def time_ten_zero(self,ind):
        self.first = ind
        self.A.remove(ind)
        self.second = self.A[0]
        self.A.remove(self.A[0])
        self.time_minutes()
            
    def time_minutes(self):
        A = self.A
        if A[1] < 6:
            self.value = 1
            if A[0] < 6:
                self.third = A[0]
                self.last = A[1]
            else:  
                self.third = A[1]
                self.last = A[0]

    def largestTimeFromDigits(self, A):
        """
        :type A: List[int]
        :rtype: str
        """
        self.value = 0
        A = sorted(A,reverse = True)
        
        self.A = A.copy()
        for number in A:
            if number == 2:
                self.time_twenty()
                if self.value == 1:
                    break
                else:
                    self.A = A.copy()
                   
            elif number == 1:
                self.time_ten_zero(1)
                if self.value == 1:
                    break
                else:
                    self.A = A.copy()
            elif number == 0: 
                self.time_ten_zero(0)
                break
            else:
                continue
                
        if self.value == 1:   
            result = [str(self.first) + str(self.second), str(self.third) + str(self.last)]
            result = ":".join(result)
            return result
        else: 
            return ""
        
A_class = Solution()
inputStr = input()
timeTemp = inputStr.split(",")
timeList = [int(num) for num in timeTemp]

output = A_class.largestTimeFromDigits(timeList)


print("output is "+ output)
```
